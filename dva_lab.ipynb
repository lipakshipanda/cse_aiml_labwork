```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Customers.csv')

# Function to show box plot
def plot_box(data, title):
    sns.boxplot(x=data)
    plt.title(title)
    plt.show()

# Visualize original
plot_box(df['Annual Income ($)'], 'Annual Income (with outliers)')

# Remove outliers using IQR
Q1, Q3 = df['Annual Income ($)'].quantile([0.25, 0.75])
IQR = Q3 - Q1
lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
df_clean = df[(df['Annual Income ($)'] >= lower) & (df['Annual Income ($)'] <= upper)]

# Visualize cleaned data
plot_box(df_clean['Annual Income ($)'], 'Annual Income (outliers removed)')

# Compare shape
print(f"Original: {df.shape}, After Removing Outliers: {df_clean.shape}")

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder

# ------------------------------
# 1. Create a sample dataset
# ------------------------------
data = {
    'Date': ['2023-01-15', '2022-05-10', '2021-12-20', '2023-07-30'],
    'Value': [10, 25, 50, 75],    # Continuous column for scaling & binning
    'Category': ['A', 'B', 'A', 'C']  # Categorical column for encoding
}

df = pd.DataFrame(data)


# ------------------------------
# 2. Extract year, month, day
# ------------------------------
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
print("With Year, Month, Day:\n", df, "\n")

# ------------------------------
# 3. Scaling (Min-Max & Standardization)
# ------------------------------
scaler_mm = MinMaxScaler()
scaler_std = StandardScaler()

df['Value_MinMax'] = scaler_mm.fit_transform(df[['Value']])
df['Value_Std'] = scaler_std.fit_transform(df[['Value']])

print("After Scaling:\n", df[['Value','Value_MinMax','Value_Std']], "\n")

# ------------------------------
# 4. Encoding
# ------------------------------

# Label Encoding
le = LabelEncoder()
df['Category_Label'] = le.fit_transform(df['Category'])
print("Label Encoding:\n", df[['Category','Category_Label']], "\n")

# One-Hot Encoding
df_onehot = pd.get_dummies(df['Category'], prefix='Category')
print("One-Hot Encoding:\n", df_onehot, "\n")

# Combine back
df = pd.concat([df, df_onehot], axis=1)

# ------------------------------
# 5. Binning (Low, Medium, High)
# ------------------------------
bins = [0, 25, 50, 100]   # ranges
labels = ['Low', 'Medium', 'High']
df['Value_Bin'] = pd.cut(df['Value'], bins=bins, labels=labels)

print("After Binning:\n", df[['Value', 'Value_Bin']], "\n")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load Wine dataset
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)

# Apply scaling
X_minmax = pd.DataFrame(MinMaxScaler().fit_transform(X), columns=X.columns)
X_std = pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns)

# Features to plot
features = ["alcohol", "malic_acid", "color_intensity"]

# Plot
fig, axes = plt.subplots(len(features), 3, figsize=(12, 8))
for i, f in enumerate(features):
    for j, data in enumerate([X, X_minmax, X_std]):
        axes[i, j].hist(data[f], bins=20, edgecolor="black")
        axes[i, j].set_title([f"(Original)", "(Min-Max)", "(Standardized)"][j])
        if j == 0: axes[i, j].set_ylabel(f)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Given array
x = np.array([[1, 2],
              [3, 4],
              [5, 6]])

print("Original Array:\n", x, "\n")

# --- Standardization ---
scaler_std = StandardScaler()
x_std = scaler_std.fit_transform(x)
print("Standardized Array (z-score):\n", x_std, "\n")

# --- Normalization (0,1) ---
scaler_mm = MinMaxScaler()
x_norm = scaler_mm.fit_transform(x)
print("Normalized Array [0,1]:\n", x_norm, "\n")

# --- Manual Hand Calculation Check for one value ---
# Let's check x[0,0] = 1 (first element)

mean_col0 = np.mean(x[:,0])   # mean of first column
std_col0 = np.std(x[:,0])     # std dev of first column
z_score = (1 - mean_col0) / std_col0

min_col0 = np.min(x[:,0])
max_col0 = np.max(x[:,0])
norm_val = (1 - min_col0) / (max_col0 - min_col0)

print("Manual Calculation for x[0,0] = 1:")
print("Standardized (z-score):", z_score)
print("Normalized (0,1):", norm_val)
```

```python
import seaborn as sns, pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load dataset
titanic = sns.load_dataset("titanic").dropna(subset=["sex", "survived"])
X, y = titanic[["sex"]], titanic["survived"]

#1
encodings = {
    "Label Encoding": pd.DataFrame({"sex": LabelEncoder().fit_transform(X["sex"])}),
    "One-Hot Encoding": pd.get_dummies(X, drop_first=True)
}

# 2
for name, X_enc in encodings.items():
    X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.2, random_state=42)
    acc = accuracy_score(y_test, LogisticRegression().fit(X_train, y_train).predict(X_test))
    print(f"Accuracy with {name}: {acc:.3f}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
dates = pd.date_range('2025-08-01', periods=30)
temps = np.random.normal(25, 3, size=30)

df = pd.DataFrame({'Date': dates, 'Temp': temps})
df['7-day MA'] = df['Temp'].rolling(7).mean()

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Temp'], label='Daily Temp', marker='o')
plt.plot(df['Date'], df['7-day MA'], label='7-day MA', color='red', linewidth=2)


max_day = df.loc[df['Temp'].idxmax()]
plt.annotate('Hottest Day\n',
             xy=(max_day['Date'], max_day['Temp']),
             xytext=(max_day['Date'], max_day['Temp']+1),
             arrowprops=dict(arrowstyle='->'))


for date in df['Date']:
    if date.weekday() >= 5:
        plt.axvspan(date, date + pd.Timedelta(days=1), color='gray', alpha=0.2)

plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.title('Daily Temperature with 7-Day Moving Average')
plt.show()
```

```python
# ==========================
# Lab 6 Solutions
# ==========================

# Install and import packages
!pip install plotly pandas numpy

import plotly.express as px
import pandas as pd
import numpy as np

# ==========================================
# Q1: Tips dataset (Plotly Express)
# ==========================================

# Load dataset
tips = px.data.tips()

# (a) Violin plot of total_bill by day, colored by time
fig1 = px.violin(
    tips,
    x="day",
    y="total_bill",
    color="time",
    box=True,
    points="all",
    title="Distribution of Total Bill by Day and Time",
    labels={"total_bill": "Total Bill ($)", "day": "Day of Week"}
)
fig1.show()

# (b) Summary statistics by day and time
summary = (
    tips.groupby(["day", "time"])["total_bill"]
    .describe(percentiles=[0.25, 0.5, 0.75])
    .rename(columns={"25%": "Q1", "50%": "Median", "75%": "Q3"})
)

summary["IQR"] = summary["Q3"] - summary["Q1"]
summary_sorted = summary.sort_values("IQR", ascending=False)[["Median", "Q1", "Q3", "IQR"]]
summary_sorted.reset_index(inplace=True)

print("\n--- Summary Statistics by Day and Time ---\n")
print(summary_sorted)


```

```python
# ==========================================
# Q2: Iris dataset (Plotly Express)
# ==========================================

iris = px.data.iris()

# (a) Pair plot of numeric features
fig2 = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Pair Plot of Iris Features"
)
fig2.show()

# (b) Pearson correlation matrix
corr = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]].corr(method="pearson")
print("\n--- Pearson Correlation Matrix ---\n")
print(corr)

# Find the pair with the largest absolute correlation
corr_unstacked = corr.unstack().drop_duplicates()
corr_unstacked = corr_unstacked[corr_unstacked.index.get_level_values(0) != corr_unstacked.index.get_level_values(1)]
max_pair = corr_unstacked.abs().idxmax()
max_corr = corr_unstacked[max_pair]

print(f"\nStrongest correlation: {max_pair[0]} and {max_pair[1]} (r = {max_corr:.3f})")

# (c) Focused scatter plot for that pair
fig3 = px.scatter(
    iris,
    x=max_pair[0],
    y=max_pair[1],
    color="species",
    trendline="ols",
    title=f"Relationship Between {max_pair[0]} and {max_pair[1]} (r = {max_corr:.3f})"
)
fig3.show()


```

```python
# ==========================================
# Q3: Gapminder dataset (Plotly Express)
# ==========================================

gapminder = px.data.gapminder()

# (a) Animated bubble chart
fig4 = px.scatter(
    gapminder,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    hover_name="country",
    animation_frame="year",
    animation_group="country",
    range_x=[100, 100000],
    range_y=[20, 90],
    log_x=True,
    title="Gapminder: GDP per Capita vs Life Expectancy (1952–2007)",
    hover_data={
        "country": True,
        "continent": True,
        "year": True,
        "gdpPercap": ":.0f",
        "lifeExp": ":.1f",
        "pop": True,
    },
)

# (b) Add dropdown for log vs linear x-axis
fig4.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="right",
            x=0.7,
            y=1.2,
            buttons=[
                dict(
                    label="Log X-axis",
                    method="relayout",
                    args=[{"xaxis.type": "log"}],
                ),
                dict(
                    label="Linear X-axis",
                    method="relayout",
                    args=[{"xaxis.type": "linear"}],
                ),
            ],
        )
    ]
)

# (d) Add annotation for insights (example: 2007)
fig4.add_annotation(
    text=(
        "In 2007, countries cluster by continent. "
        "Africa shows lower GDP and life expectancy, "
        "while Europe and Oceania are higher on both scales."
    ),
    xref="paper",
    yref="paper",
    x=0.5,
    y=-0.25,
    showarrow=False,
)

fig4.show()

# (e) Save figure as standalone HTML (with legend interactivity)
fig4.write_html("gapminder_animation.html")
print("\nAnimated Gapminder chart saved as gapminder_animation.html")
```
